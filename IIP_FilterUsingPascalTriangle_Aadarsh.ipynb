Convolution with kernel from pascal triangle

```python
import cv2
import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
```

cv2.filter2D function is used to apply convolution to the image using the specified kernel. The kernel variable holds the 3x3 Gaussian blur kernel derived from Pascal's triangle. The -1 argument in the function specifies that the output image should have the same depth as the input image.

```python
# Loading an image using OpenCV
image = cv2.imread('/content/flower1.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Define a kernel (3x3 Gaussian blur kernel from Pascal's triangle)
kernel = np.array([[1, 2, 1],
                   [2, 4, 2],
                   [1, 2, 1]]) / 16
```

```python
# Apply convolution using OpenCV's filter2D function
filtered_image_opencv = cv2.filter2D(image, -1, kernel)
```

```python
# Display the original and filtered images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap="gray")
```

```python
plt.subplot(1, 2, 2)
plt.title("Filtered Image (OpenCV)")
plt.imshow(filtered_image_opencv, cmap="gray")

plt.tight_layout()
plt.show()
```

```python
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]]) / 1
```

```python
sharpened_image = cv2.filter2D(image, -1, sharpen_kernel)
```

```python
plt.subplot(1, 2, 2)
plt.title("Sharpened Image (OpenCV)")
plt.imshow(sharpened_image, cmap="gray")

plt.tight_layout()
plt.show()
```
