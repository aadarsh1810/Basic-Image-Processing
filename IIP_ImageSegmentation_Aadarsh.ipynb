```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
image = cv2.imread('/content/flower1.jpg', cv2.IMREAD_GRAYSCALE)
new_width = 500
new_height = 500

# Resize the image
resized_image = cv2.resize(image, (new_width, new_height))
cv2_imshow(resized_image)
```

```python
if image is None:
    print("Image not found.")
else:
    # Apply different thresholding techniques

    # Binary Thresholding
    _, binary_thresholded = cv2.threshold(resized_image, 128, 255, cv2.THRESH_BINARY)

    # Truncation Thresholding
    _, truncation_thresholded = cv2.threshold(resized_image, 128, 255, cv2.THRESH_TRUNC)
cv2_imshow( binary_thresholded)
cv2_imshow(truncation_thresholded)
#cv2_imshow('Adaptive Thresholding', adaptive_thresholded)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Apply segmentation techniques
threshold_value = 128  # Adjust this value as needed

# THRESH_BINARY_INV
ret, thresh_binary_inv = cv2.threshold(resized_image, threshold_value, 255, cv2.THRESH_BINARY_INV)

# THRESH_TOZERO
ret, thresh_tozero = cv2.threshold(resized_image, threshold_value, 255, cv2.THRESH_TOZERO)

# THRESH_TOZERO_INV
ret, thresh_tozero_inv = cv2.threshold(resized_image, threshold_value, 255, cv2.THRESH_TOZERO_INV)

# Display the segmented images
cv2_imshow(thresh_binary_inv)
cv2_imshow(thresh_tozero)
cv2_imshow( thresh_tozero_inv)

# Wait for a key press and then close all windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```
