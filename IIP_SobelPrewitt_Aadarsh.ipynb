```python
import cv2
import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
```

```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/flower1.jpg')
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Applying Gaussian blur filter
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
# Increasing the kernel size will increase the blurness
cv2_imshow(blurred_image)
```

```python
#for horizontal edges
gray_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2GRAY)

sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)

magnitude_h = cv2.magnitude(sobel_x, sobel_y)
cv2_imshow( magnitude_h)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#for vertical edges
magnitude_v = cv2.magnitude(sobel_y, sobel_x)
cv2_imshow(magnitude_v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_image)
```

```python
prewitt_x = cv2.filter2D(gray_image, cv2.CV_64F, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
prewitt_y = cv2.filter2D(gray_image, cv2.CV_64F, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

magnitude_x = cv2.magnitude(prewitt_x, prewitt_y)
magnitude_y = cv2.magnitude(prewitt_y, prewitt_x)

cv2_imshow( magnitude_x)#horizontal
cv2_imshow( magnitude_y)#vertical
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Creating a synthetic image with vertical and horizontal edges
matrix = np.array([[20, 20, 200],
                   [20,20,200],
                   [20, 20, 200]], dtype=np.uint8)

# Displaying the matrix
print("Matrix:")
print(matrix)

# Creating a synthetic grayscale image from the matrix
synthetic_image = cv2.resize(matrix, (300, 300), interpolation=cv2.INTER_NEAREST)

# Displaying the synthetic image
cv2_imshow(synthetic_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Applying Sobel edge detection
sobel_x = cv2.Sobel(synthetic_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(synthetic_image, cv2.CV_64F, 0, 1, ksize=3)

magnitude_sobel_x = cv2.magnitude(sobel_x, sobel_y)
cv2_imshow( magnitude_sobel_x)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Applying Prewitt edge detection
prewitt_x = cv2.filter2D(synthetic_image, cv2.CV_64F, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
prewitt_y = cv2.filter2D(synthetic_image, cv2.CV_64F, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

magnitude_prewitt_x = cv2.magnitude(prewitt_x, prewitt_y)
cv2_imshow( magnitude_prewitt_x)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
